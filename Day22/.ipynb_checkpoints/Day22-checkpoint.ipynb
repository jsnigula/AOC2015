```python
state = (True, '', 50, 51, 500, 0, 0, 0, 0)
    
states = [state]
    
best_cost = 1e9
best_fight = ''
    
done = False
lvl = 1

while not done:
    
    nextstates = []

    for s in states:
        pturn, fight, hp, bhp, mana, c_s, c_p, c_r, cost = s

        if bhp < 1:
            if cost < best_cost:
                best_cost = cost
                best_fight = fight
                # print('Best cost now', best_cost, best_fight)
            continue

        if cost > best_cost:
            continue
        # Resolve ongoing effects
        bd = 9
        if c_s != 0:
            bd = 2 # 7 damage less
        if c_p != 0:
            bhp -= 3
        if c_r != 0:
            mana += 101

        if bhp < 1:
            if cost < best_cost:
                best_cost = cost
                best_fight = fight
                # print('Best cost now', best_cost, best_fight)
            continue

        c_s = max(c_s-1, 0)
        c_p = max(c_p-1, 0)
        c_r = max(c_r-1, 0)
    
        if pturn:        

            # Player turn
            #  print('Player turn')
            if mana < 53:
                continue
        
            if mana >= 53:
                nextstates.append((not pturn, fight+f'M_{hp}-{bhp-4}', hp, bhp-4, mana-53, c_s, c_p, c_r, cost+53))
                    
            if mana >= 73:
                nextstates.append((not pturn, fight+f'D_{min(hp+2, 50)}-{bhp-2}', min(hp+2, 50), bhp-2, mana-73, c_s, c_p, c_r, cost+73))

            if mana >= 113 and c_s == 0:
                nextstates.append((not pturn, fight+f'S_{hp}-{bhp}', hp, bhp, mana-113, 6, c_p, c_r, cost+113))

            if mana >= 173 and c_p == 0:
                nextstates.append((not pturn, fight+f'P_{hp}-{bhp}', hp, bhp, mana-173, c_s, 6, c_r, cost+173))

            if mana >= 229 and c_r == 0:
                nextstates.append((not pturn, fight+f'R_{hp}-{bhp}', hp, bhp, mana-229, c_s, c_p, 5, cost+229))
            
        else:
            hp -= bd
            if hp - bd < 1:
                continue
                
            nextstates.append((not pturn, fight+f'B_{hp}-{bhp}', hp, bhp, mana, c_s, c_p, c_r, cost))
            
    states = nextstates
    if len(states) == 0:
        done = True
    print('Level', lvl, 'evaluating', len(states), 'states')
    lvl += 1
        
print(best_cost)
print(best_fight)
```
